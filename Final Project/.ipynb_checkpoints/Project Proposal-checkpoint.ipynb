```python
"""
Crawl Ranked Solo/Duo (queue=420) teammate edges for the last 30 days starting from a Riot ID,
then expand one hop (teammates' teammates). Collect winrate stats and build an edge table.

- Start: Mooncake#2003
- Region routing (match/account): americas
- Platform routing (summoner/league endpoints): na1  (not strictly needed here)

Outputs:
  data/edges.csv
  data/players.csv

DUO EXCLUSION:
  1) If match-v5 participants include 'partyId' (or other premade key), we remove edges between duo partners strictly.
  2) Otherwise, we use a heuristic: if two players appear as teammates >= DUO_REPEAT_THRESHOLD within the window,
     mark them as likely-duo and exclude those edges. (You can disable heuristic.)

Notes:
- Requires: pip install requests pandas python-dateutil
- Set your API key in env RIOT_API_KEY or replace API_KEY below.
"""

import os
import time
import math
import json
import random
from dataclasses import dataclass
from datetime import datetime, timedelta, timezone
from typing import Dict, List, Tuple, Set, Optional, Any

import requests
import pandas as pd

# -----------------------
# Config
# -----------------------
API_KEY = os.getenv("RIOT_API_KEY", "").strip()
if not API_KEY:
    raise RuntimeError("Missing RIOT_API_KEY env var. Please export RIOT_API_KEY=YOUR_KEY")

START_GAME_NAME = "Mooncake"
START_TAG_LINE  = "0208"

REGION_ROUTING = "americas"   # for account-v1 + match-v5 in NA
QUEUE_ID = 420               # Ranked Solo/Duo
DAYS_BACK = 30
MAX_MATCHES_PER_PLAYER = 1000  # safety cap; matchlist endpoint returns newest first

# expansion
MAX_TEAMMATES_SEED = 300      # cap number of unique teammates from seed player to expand
MAX_EXPANDED_PLAYERS = 300    # cap number of teammates whose own matchlists we crawl

# rate limiting
BASE_SLEEP = 0.12
JITTER = 0.08

# duo exclusion
ENABLE_HEURISTIC_DUO_FILTER = True
DUO_REPEAT_THRESHOLD = 6  # if two players co-team >= this many times in 30d, likely duo/premade
# -----------------------

HEADERS = {"X-Riot-Token": API_KEY}

def unix_now() -> int:
    return int(time.time())

def unix_days_ago(days: int) -> int:
    return int((datetime.now(timezone.utc) - timedelta(days=days)).timestamp())

def sleep_jitter(mult: float = 1.0):
    time.sleep(max(0.0, BASE_SLEEP * mult + random.random() * JITTER))

class RiotHTTPError(RuntimeError):
    pass

def riot_get(url: str, params: Optional[dict] = None, max_retries: int = 8) -> Any:
    """
    Riot GET with basic retry + 429 handling.
    """
    for attempt in range(max_retries):
        r = requests.get(url, headers=HEADERS, params=params, timeout=30)
        if r.status_code == 200:
            sleep_jitter()
            return r.json()

        if r.status_code == 429:
            retry_after = r.headers.get("Retry-After")
            wait = float(retry_after) if retry_after else (1.5 + attempt * 0.8)
            time.sleep(wait)
            continue

        if 500 <= r.status_code < 600:
            time.sleep(1.0 + attempt * 0.6)
            continue

        # hard fail
        raise RiotHTTPError(f"HTTP {r.status_code} for {url}: {r.text[:300]}")

    raise RiotHTTPError(f"Max retries exceeded for {url}")

# -----------------------
# Riot endpoints
# -----------------------
def account_by_riot_id(game_name: str, tag_line: str) -> dict:
    url = f"https://{REGION_ROUTING}.api.riotgames.com/riot/account/v1/accounts/by-riot-id/{game_name}/{tag_line}"
    return riot_get(url)

def match_ids_by_puuid(puuid: str, start_time_unix: int, queue: int, count: int = 100) -> List[str]:
    """
    Pull match IDs for puuid. We paginate with 'start' until no more.
    """
    all_ids = []
    start = 0
    while True:
        url = f"https://{REGION_ROUTING}.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids"
        params = {
            "queue": queue,
            "startTime": start_time_unix,
            "start": start,
            "count": min(count, 100),
        }
        ids = riot_get(url, params=params)
        if not ids:
            break
        all_ids.extend(ids)
        if len(ids) < params["count"]:
            break
        start += params["count"]
        if len(all_ids) >= MAX_MATCHES_PER_PLAYER:
            break
    return all_ids[:MAX_MATCHES_PER_PLAYER]

def match_detail(match_id: str) -> dict:
    url = f"https://{REGION_ROUTING}.api.riotgames.com/lol/match/v5/matches/{match_id}"
    return riot_get(url)

# -----------------------
# Parsing utilities
# -----------------------
def extract_team_edges_from_match(md: dict) -> Tuple[List[dict], List[dict]]:
    """
    From match detail, build:
      - edges: for each player, edges to each teammate (directed or undirected)
      - players rows: per participant stats for winrate aggregation

    Edge record fields:
      src_puuid, dst_puuid, match_id, game_start, queue_id, team_id, win (src team)
      plus optional: party_id_src, party_id_dst (if present)
    """
    info = md.get("info", {})
    match_id = md.get("metadata", {}).get("matchId")
    queue_id = info.get("queueId")
    game_start = info.get("gameStartTimestamp")  # ms
    participants = info.get("participants", [])

    # Build lookups by team
    team_to_puuids: Dict[int, List[dict]] = {}
    player_rows: List[dict] = []
    edges: List[dict] = []

    for p in participants:
        puuid = p.get("puuid")
        team_id = int(p.get("teamId"))
        win = bool(p.get("win"))
        # potential premade indicator (may not exist)
        party_id = p.get("partyId") or p.get("party") or p.get("party_id")  # defensive
        team_to_puuids.setdefault(team_id, []).append({
            "puuid": puuid,
            "win": win,
            "partyId": party_id,
        })
        player_rows.append({
            "puuid": puuid,
            "matchId": match_id,
            "queueId": queue_id,
            "gameStartTimestamp": game_start,
            "teamId": team_id,
            "win": win,
        })

    # Create teammate co-play edges (directed src->dst, excluding self)
    for team_id, plist in team_to_puuids.items():
        for a in plist:
            for b in plist:
                if a["puuid"] == b["puuid"]:
                    continue
                edges.append({
                    "src_puuid": a["puuid"],
                    "dst_puuid": b["puuid"],
                    "matchId": match_id,
                    "queueId": queue_id,
                    "gameStartTimestamp": game_start,
                    "teamId": team_id,
                    "src_win": a["win"],
                    "src_partyId": a.get("partyId"),
                    "dst_partyId": b.get("partyId"),
                })

    return edges, player_rows

def strict_duo_edge_filter(edges: List[dict]) -> Tuple[List[dict], bool]:
    """
    If partyId exists (non-null) for both endpoints and equal, treat as premade and remove.
    Returns filtered_edges, used_strict_flag
    """
    any_party = any(e.get("src_partyId") is not None or e.get("dst_partyId") is not None for e in edges)
    if not any_party:
        return edges, False

    filtered = []
    for e in edges:
        a = e.get("src_partyId")
        b = e.get("dst_partyId")
        # if both known and identical => same premade group (duo/trio/etc)
        if a is not None and b is not None and a == b:
            continue
        filtered.append(e)
    return filtered, True

def build_repeat_pair_counts(edges: List[dict]) -> Dict[Tuple[str, str], int]:
    """
    Count undirected co-team repeats for pair across the window.
    """
    counts: Dict[Tuple[str, str], int] = {}
    for e in edges:
        u, v = e["src_puuid"], e["dst_puuid"]
        if u == v:
            continue
        a, b = (u, v) if u < v else (v, u)
        counts[(a, b)] = counts.get((a, b), 0) + 1
    # because edges are directed and we add both directions within team,
    # each match contributes 2 to the pair count. Convert to matches count:
    for k in list(counts.keys()):
        counts[k] = counts[k] // 2
    return counts

def heuristic_duo_edge_filter(edges: List[dict], threshold_matches: int) -> List[dict]:
    """
    Remove edges for pairs that repeat >= threshold_matches in the window.
    This is a proxy for duo/premade when partyId is unavailable.
    """
    pair_counts = build_repeat_pair_counts(edges)
    duo_pairs: Set[Tuple[str, str]] = {k for k, c in pair_counts.items() if c >= threshold_matches}

    filtered = []
    for e in edges:
        u, v = e["src_puuid"], e["dst_puuid"]
        a, b = (u, v) if u < v else (v, u)
        if (a, b) in duo_pairs:
            continue
        filtered.append(e)
    return filtered

# -----------------------
# Main crawl
# -----------------------
def crawl_player_window(puuid: str, start_time_unix: int) -> Tuple[List[dict], List[dict]]:
    """
    Crawl matches for one player in the window; return (edges, player_rows).
    """
    ids = match_ids_by_puuid(puuid, start_time_unix=start_time_unix, queue=QUEUE_ID, count=100)
    all_edges: List[dict] = []
    all_players: List[dict] = []

    for mid in ids:
        md = match_detail(mid)
        edges, players = extract_team_edges_from_match(md)
        all_edges.extend(edges)
        all_players.extend(players)

    return all_edges, all_players

def main():
    os.makedirs("data", exist_ok=True)
    start_time_unix = unix_days_ago(DAYS_BACK)

    # 0) seed puuid
    acct = account_by_riot_id(START_GAME_NAME, START_TAG_LINE)
    seed_puuid = acct["puuid"]
    print(f"[seed] {START_GAME_NAME}#{START_TAG_LINE} puuid={seed_puuid}")

    # 1) crawl seed
    seed_edges, seed_players = crawl_player_window(seed_puuid, start_time_unix)
    print(f"[seed] raw edges={len(seed_edges)} player_rows={len(seed_players)}")

    # strict duo removal if partyId exists
    seed_edges, used_strict = strict_duo_edge_filter(seed_edges)
    print(f"[seed] strict_duo_filter_used={used_strict} edges_after_strict={len(seed_edges)}")

    # 2) teammates from seed (excluding self)
    seed_teammates = sorted({e["dst_puuid"] for e in seed_edges if e["src_puuid"] == seed_puuid})
    if len(seed_teammates) > MAX_TEAMMATES_SEED:
        seed_teammates = seed_teammates[:MAX_TEAMMATES_SEED]
    print(f"[seed] unique teammates (post-strict-filter) = {len(seed_teammates)}")

    # 3) expand one hop: crawl teammates
    expanded_edges_all: List[dict] = []
    expanded_players_all: List[dict] = []

    expanded_list = seed_teammates[:MAX_EXPANDED_PLAYERS]
    for idx, t_puuid in enumerate(expanded_list, 1):
        try:
            e2, p2 = crawl_player_window(t_puuid, start_time_unix)
            # apply strict duo filter if possible (match-level partyId presence)
            e2, _ = strict_duo_edge_filter(e2)
            expanded_edges_all.extend(e2)
            expanded_players_all.extend(p2)
            if idx % 10 == 0:
                print(f"[expand] {idx}/{len(expanded_list)} crawled, edges={len(expanded_edges_all)}")
        except RiotHTTPError as ex:
            print(f"[expand][warn] failed for teammate {t_puuid}: {ex}")
            continue

    # 4) combine edges + player rows
    all_edges = seed_edges + expanded_edges_all
    all_players = seed_players + expanded_players_all

    print(f"[all] combined edges={len(all_edges)} player_rows={len(all_players)}")

    # 5) If partyId wasn't available, apply heuristic duo filter (optional)
    if not used_strict and ENABLE_HEURISTIC_DUO_FILTER:
        before = len(all_edges)
        all_edges = heuristic_duo_edge_filter(all_edges, threshold_matches=DUO_REPEAT_THRESHOLD)
        print(f"[all] heuristic_duo_filter applied: {before} -> {len(all_edges)} edges")

    # 6) build winrate stats per player from player_rows
    dfp = pd.DataFrame(all_players)
    # Keep only queue=420 rows (defensive)
    dfp = dfp[dfp["queueId"] == QUEUE_ID].copy()

    stats = (
        dfp.groupby("puuid")["win"]
           .agg(games="count", wins="sum")
           .reset_index()
    )
    stats["winrate"] = stats["wins"] / stats["games"]

    # 7) edge table
    dfe = pd.DataFrame(all_edges)
    # keep only queue=420 (defensive)
    dfe = dfe[dfe["queueId"] == QUEUE_ID].copy()

    # 8) (Optional) add helpful columns: date
    if "gameStartTimestamp" in dfe.columns:
        dfe["gameStartUTC"] = pd.to_datetime(dfe["gameStartTimestamp"], unit="ms", utc=True)

    # 9) save
    edges_path = "data/edges.csv"
    players_path = "data/players.csv"
    dfe.to_csv(edges_path, index=False)
    stats.to_csv(players_path, index=False)

    print(f"[done] wrote {edges_path} rows={len(dfe)}")
    print(f"[done] wrote {players_path} rows={len(stats)}")

if __name__ == "__main__":
    main()
```
