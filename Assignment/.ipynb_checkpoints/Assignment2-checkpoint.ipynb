## Question 1

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community import louvain_communities
from networkx.algorithms.community.quality import modularity
from scipy.stats import gaussian_kde
from scipy.io import mmread
import math
```

```python
G1 = nx.read_adjlist("mystery_edges1.txt", nodetype=int)
G2 = nx.read_adjlist("mystery_edges2.txt", nodetype=int)
```

In the BA model, we can calculate the total number of edges given n and m using the following formula. Similarly, I can deduce the value of m from the total number of edges and n.
$$E(n) = \frac{m_0 (m_0 - 1)}{2} + m (n - m_0)$$
However, for common cases, we usually define m0 = m, and obtain the following formula.
$$E(n) = mn - \frac{m(m+1)}{2}$$
Then, by rearranging the equation, we can obtain a quadratic equation about m.
$$ m^2 + (1-2n)m + 2E = 0$$

```python
n1 = G1.number_of_nodes()
n2 = G2.number_of_nodes()
E1 = G1.number_of_edges()
E2 = G2.number_of_edges()

def cal_m(n, E):
    b = 1 - 2*n
    c = 2*E

    discriminant = b**2 - 4*c
    m1 = (-b + math.sqrt(discriminant)) / 2
    m2 = (-b - math.sqrt(discriminant)) / 2

    return m1,m2

G1_m1, G1_m2 = cal_m(n1,E1)
G2_m1, G2_m2 = cal_m(n2,E2)
print(min(G1_m1,G1_m2))
print(min(G2_m1,G2_m2))
```

```python
def modularity_score(G, seed=0):
    comms = louvain_communities(G, seed=seed)
    return nx.algorithms.community.quality.modularity(G, comms)
    
def graph_stats(G, seed=0):
    c = nx.average_clustering(G)
    r = nx.degree_assortativity_coefficient(G)
    q = modularity_score(G, seed=seed)
    
    return c, r, q

C1, R1, Q1 = [], [], []
C2, R2, Q2 = [], [], []
for i in range(100):
    sim_G1 = nx.barabasi_albert_graph(n1,3,seed = i)
    sim_G2 = nx.barabasi_albert_graph(n2,3,seed = i)
    c1, r1, q1 = graph_stats(sim_G1, seed=i)
    c2, r2, q2 = graph_stats(sim_G2, seed=i)
    C1.append(c1)
    R1.append(r1)
    Q1.append(q1)
    C2.append(c2) 
    R2.append(r2)
    Q2.append(q2)

def plot_kde_two_ensembles(x1, x2, base1, base2, title, xlabel):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    
    x_min1 = min(x1.min(), base1)
    x_max1 = max(x1.max(), base1)
    xs1 = np.linspace(x_min1, x_max1, 500)
    kde1 = gaussian_kde(x1)
    axes[0].plot(xs1, kde1(xs1), label="BA model(m=3) KDE")
    axes[0].axvline(base1, linestyle="--", linewidth=2, label="G1")
    axes[0].set_title("G1 vs BA model(m=3)")
    axes[0].set_xlabel(xlabel)
    axes[0].set_ylabel("Density")
    axes[0].legend()

    x_min2 = min(x2.min(), base2)
    x_max2 = max(x2.max(), base2)
    xs2 = np.linspace(x_min2, x_max2, 500)

    kde2 = gaussian_kde(x2)
    axes[1].plot(xs2, kde2(xs2), label="BA model(m=3) KDE")
    axes[1].axvline(base2, linestyle="--", linewidth=2, label="G2")
    axes[1].set_title("G2 vs BA model(m=3)")
    axes[1].set_xlabel(xlabel)
    axes[1].legend()

    fig.suptitle(title)
    plt.tight_layout()
    plt.show()
    
baseC1, baseR1, baseQ1 = graph_stats(G1)
baseC2, baseR2, baseQ2 = graph_stats(G2)

plot_kde_two_ensembles(np.array(C1), np.array(C2), baseC1, baseC2,
                       title="KDE: Average Clustering",
                       xlabel="Average clustering")

plot_kde_two_ensembles(np.array(R1), np.array(R2), baseR1, baseR2,
                       title="KDE: Degree Assortativity",
                       xlabel="Assortativity")

plot_kde_two_ensembles(np.array(Q1), np.array(Q2), baseQ1, baseQ2,
                       title="KDE: Modularity",
                       xlabel="Modularity")
```

 For G1, we are quite certain that it was generated by the BA growth model. This is not only because its calculated m value is closer to an integer and the same edge number appears when constructing a BA model with the same n and m, but also because its graph features are closer to those of the BA model. However, G2 is very likely not a BA model, as its characteristics differ significantly from those of the BA model.

## Question 2

```python
ms = [1,2,3,4,5]
Ss = [50,100,150,200,250,300,350]
modular = {m: [] for m in ms}
times = 10

for m in ms:
    for s in Ss:
        p = 2 * m / s
        qs = []
        for _ in range(times):
            G = nx.erdos_renyi_graph(s, p)
            comms = louvain_communities(G)
            q = nx.algorithms.community.quality.modularity(G, comms)
            qs.append(q)

        modular[m].append(np.mean(qs))

```

```python
plt.figure(figsize=(7,5))

markers = ['o','s','D','^','<']

for i, m in enumerate(ms):
    plt.plot(Ss, modular[m], marker=markers[i], label=f"m = {m}")

plt.xlabel("Size, S")
plt.ylabel("Modularity")
plt.title("Modularity of ER Graphs vs Network Size")
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

My code performed 10 repeated measurements on the modularity of an ER random graph and took the average. The graph shows that even with low graph density, the random graph exhibits clear community characteristics. This indicates that simply observing the modularity of a target graph is insufficient to accurately determine whether it possesses community characteristics, especially when the graph with lower m. On the other hand, this also reflects that the algorithm overemphasizes max(Q) when calculating comms like louvain, leading to many communities that, while numerically represented, are actually meaningless
